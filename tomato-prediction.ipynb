# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
```

```python
model=rfr.fit(X_train, y_train)
print (model)
```

```python
rfr_y_pred = rfr.predict(X_test)
print (rfr_y_pred)
```

```python
print (type(rfr_y_pred))
```

```python
# print (type(rfr_y_pred),type(y_test))  
# print (len(rfr_y_pred),len(y_test))  
# print (rfr_y_pred.shape,y_test.shape)  
from sklearn import metrics  
import numpy as np  
sum_mean=0  
for i in range(len(rfr_y_pred)):  
    sum_mean+=(rfr_y_pred[i]-y_test.values[i])**2  
sum_erro=np.sqrt(sum_mean/50)  
# calculate RMSE by hand  
print ("RMSE of RandomForestRegressor:",sum_erro) 
```

```python
import matplotlib.pyplot as plt  
plt.figure()  
plt.plot(range(len(rfr_y_pred)),rfr_y_pred,'b',label="predict")  
plt.plot(range(len(rfr_y_pred)),y_test,'r',label="test")  
plt.legend(loc="upper right") 
plt.title("Random Forest Predict")
plt.xlabel("the number of y")  
plt.ylabel('value of y')  
plt.show()
```

# cross validation

```python
from sklearn import *
from sklearn.model_selection import cross_val_predict
# from sklearn.cross_validation import cross_val_score
predicted = cross_val_predict(rfr, X, y, cv=10)
score = cross_val_score(rfr, X_train, y_train, cv = 10).mean()
print(score)
# 用scikit-learn计算MSE
print ("MSE:",metrics.mean_squared_error(y, predicted))
# 用scikit-learn计算RMSE
print ("RMSE:", np.sqrt(metrics.mean_squared_error(y, predicted)))
```

```python
fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_title('cross validation')
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
